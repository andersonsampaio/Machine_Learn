```python
import tensorflow as tf
```

```python
a = tf.constant([[-1.0, 7.0, 5.0]], name = 'entradas')
b = tf.constant([[0.8, 0.1, 0.0]], name = 'pesos')
multiplicacao = tf.multiply(a, b)
soma = tf.reduce_sum(multiplicacao)
```

```python
print(a)
print('\n')
print(b)
print('\n')
print(multiplicacao)
print('\n')
print(soma)
```
